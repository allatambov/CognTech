### Задание 2

Так как функция распределения определяется неоднозначно, воспользуйтесь определением из предыдущей задачи, т.е. с нестрогим неравенством.

*Пример 1*

    X = np.array([2, 3, 7])
    P = np.array([0.3, 0.1, 0.6])

    F(X, P, 4.5)
    0.4

*Пример 2*

    X = np.array([-2, 3, 7, 10])
    P = np.array([0.1, 0.3, 0.1, 0.5])

    F(X, P, 7)
    0.5
